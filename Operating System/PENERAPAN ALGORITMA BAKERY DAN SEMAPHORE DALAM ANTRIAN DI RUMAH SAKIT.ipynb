### PENERAPAN ALGORITMA BAKERY DAN SEMAPHORE DALAM ANTRIAN DI RUMAH SAKIT

Salah satu contoh penggunaan algoritma bakery dengan semaphore pada kehidupan sehari-hari adalah pada sistem pengurutan antrian di sebuah layanan publik, misalnya di Rumah Sakit. Algoritma bakery dapat digunakan untuk mengatur urutan kedatangan pasien yang ingin dilayani. 

Dalam sistem ini, setiap pasien akan diberikan nomor urutan sesuai dengan urutan kedatangan mereka. Nomor urutan tersebut akan digunakan untuk menentukan urutan pasien yang akan dilayani selanjutnya. Algoritma bakery dapat digunakan untuk mengatur urutan ini dengan aman, sehingga tidak terjadi konflik antara pasien yang ingin dilayani.

Contoh penerapan algoritma bakery dengan semaphore pada sistem antrian adalah sebagai berikut:
- Terdapat kunjungan rutin pasien di sebuah Rumah Sakit sehingga diperlukan antrian untuk dapat menangani mereka satu per satu.
- Setiap pasien yang datang akan diberikan nomor urutan berdasarkan urutan kedatangan mereka.
- Setiap pasien akan memasuki sebuah antrian untuk menunggu giliran dilayani.
- Ketika sebuah pasien telah selesai dilayani, nomor urutan pasien tersebut akan dihapus dari daftar urutan.
- Algoritma bakery digunakan untuk mengatur urutan kedatangan pasien yang ingin dilayani. Dalam hal ini, setiap pasien akan melakukan proses bakery untuk menentukan nomor urutan mereka dalam antrian.
- Dalam bagian kritis, hanya satu pasien yang akan dilayani pada satu waktu. Algoritma bakery akan memastikan bahwa pasien yang memiliki nomor urutan lebih kecil akan dilayani terlebih dahulu sebelum pasien dengan nomor urutan yang lebih besar.
- Namun, dalam situasi tertentu mungkin terjadi sebuah kondisi dimana dalam prosesnya. Misalnya, ketika dua atau lebih pasien saling memblokir untuk masuk ke ruangan dokter.
    - Pasien A mendapatkan nomor urutan 1 dan masuk ke ruangan dokter, tetapi dia tidak keluar dari ruangan.
    - Pasien B mendapatkan nomor urutan 2 dan menunggu Pasien A keluar dari ruangan, tetapi Pasien A tidak keluar.
    - Pasien C mendapatkan nomor urutan 3 dan menunggu Pasien B keluar dari ruangan, tetapi Pasien B tidak keluar karena masih menunggu Pasien A keluar.
    - Demikian seterusnya. Oleh karena itu diperlukan solusi dalam menyeelesaikan masalah ini yaitu dengan menggunakan semaphore.
- Semaphore memungkinkan penggunaan bersama sumber daya dengan membatasi jumlah proses yang dapat mengaksesnya secara bersamaan sehingga proses yang tertunda dapat segera diatasi.
- Selanjutnya, jika pasien sudah selesai dilayani, maka pasien akan keluar dari antrian dan nomor urutan mereka akan dihapus dari daftar urutan.
- Dengan menggunakan algoritma bakery dalam sistem antrian, pengaturan urutan kedatangan pasien menjadi lebih aman dan efektif. Hal ini dapat meminimalkan konflik antara pasien yang ingin dilayani dan memastikan bahwa setiap pasien akan dilayani sesuai dengan urutan kedatangan mereka. 

```python
from threading import Thread, Semaphore
import time
import random

# Inisialisasi variabel global
n = 1  # penambahan untuk indeks
max_antrian = 10  # Batas maksimum antrian
waktu_tunggu = []  # Menyimpan waktu tunggu pasien
nama_pasien = []  # Menyimpan nama pasien

# Semaphore untuk mengendalikan akses ke bagian kritis
semaphore = Semaphore()

def maxNumber():
    # Fungsi untuk mencari nomor urutan terbesar
    max_num = antrian[0]
    for i in range(1, len(antrian)):
        if antrian[i] > max_num:
            max_num = antrian[i]
    return max_num

def bakery(i):
    global entering, antrian, semaphore
    entering[i] = True
    antrian[i] = maxNumber() + 1
    entering[i] = False

    for j in range(len(antrian)):
        while entering[j]:
            pass  # Menunggu proses dengan nomor urutan yang lebih kecil selesai memasuki antrian
        while antrian[j] != 0 and (antrian[j] < antrian[i] or (antrian[j] == antrian[i] and j < i)):
            pass
        '''Menunggu proses dengan nomor urutan yang lebih kecil atau dengan nomor urutan 
        yang sama tetapi memiliki indeks yang lebih kecil selesai memasuki antrian'''

    # Bagian ini merupakan bagian kritis yang harus dilindungi oleh algoritma bakery
    # Proses pelayanan
    print(f'Pasien {i + n} ({nama_pasien[i]}) memasuki ruangan dokter...')

    # Mengakuisisi semaphore untuk masuk ke bagian kritis
    semaphore.acquire()

    print(f'Pasien {i + n} ({nama_pasien[i]}) sedang dalam proses pelayanan...')
    time.sleep(random.randint(1, 20))  # Proses pelayanan

    # Melepaskan semaphore setelah selesai dari bagian kritis
    semaphore.release()

    print(f'Pasien {i + n} ({nama_pasien[i]}) sudah selesai dari ruangan dokter...')
    print("-------------------------------------------------")

    # Mengecek antrian selanjutnya
    if i + 1 < len(antrian) and antrian[i + 1] > 0:
        print(f'Pasien {i + (n + 1)} ({nama_pasien[i + 1]}) selanjutnya')
        print("-------------------------------------------------")

    # Keluar dari antrian
    antrian[i] = 0

def tambahAntrian(nama): # fungsi untuk menambahkan proses pada antrian
    global nama_pasien, waktu_tunggu, entering, antrian

    if len(antrian) < max_antrian:
        print(f"Antrian Pasien {nama}")
        print("---------------------")
        nama_pasien.append(nama)
        waktu_tunggu.append(time.time())  # Waktu pasien bergabung ke antrian
        entering.append(False)
        antrian.append(0)
    else:
        print("Batas maksimum antrian telah tercapai.")

def batalkanAntrian(): # fungsi untuk membatalkan pasien yang sudah masuk dalam antrian
    global nama_pasien, waktu_tunggu, entering, antrian

    if len(antrian) > 0:
        index = int(input("Masukkan nomor antrian yang ingin dibatalkan: "))
        if 1 <= index <= len(antrian):
            index -= 1
            print(f"Antrian Pasien {nama_pasien[index]} dibatalkan.")
            del nama_pasien[index]
            del waktu_tunggu[index]
            del entering[index]
            del antrian[index]
        else:
            print("Nomor antrian tidak valid.")
    else:
        print("Antrian kosong.")

def print_status(): # fungsi untuk mencetak status antrian
    global nama_pasien, waktu_tunggu, antrian

    if len(antrian) > 0:
        print("Status Antrian:")
        for i in range(len(antrian)):
            print(f"Nomor Antrian: {i + 1}")
            print(f"Nama Pasien: {nama_pasien[i]}")
            print("------------------------------")
    else:
        print("Antrian kosong.")

def main():
    global entering, antrian

    entering = []  # Initialize entering list
    antrian = []  # Initialize antrian list

    while True:
        print("==============================================")
        print("| SELAMAT DATANG DI MENU ANTRIAN RUMAH SAKIT |")
        print("==================== Menu ====================")
        print("| 1. Antrian Pasien                          |")
        print("| 2. Tambah Antrian                          |")
        print("| 3. Batalkan Antrian                        |")
        print("| 4. Cek Status Antrian                      |")
        print("| 5. Keluar                                  |")
        print("==============================================")
        choice = input("Pilih menu: ")
        print("----------------------------------------------\n")

        if choice == "1":
            num_processes = len(antrian)
            entering = [False] * num_processes
            antrian = [0] * num_processes  # Initialize antrian list
            threads = []

            if num_processes == 0:
                print("Antrian kosong.\n")
            else:
                for i in range(num_processes):
                    threads.append(Thread(target=bakery, args=(i,)))

                for thread in threads:
                    thread.start()

                for thread in threads:
                    thread.join()

        elif choice == "2":
            nama = input("Masukkan nama pasien: ")
            tambahAntrian(nama)
        elif choice == "3":
            batalkanAntrian()
        elif choice == "4":
            print_status()
        elif choice == "5":
            print("Antrian Selesai, Terimakasih.")
            break
        else:
            print("Pilihan tidak valid. Silakan pilih lagi.")

if __name__ == "__main__":
    main()
```
